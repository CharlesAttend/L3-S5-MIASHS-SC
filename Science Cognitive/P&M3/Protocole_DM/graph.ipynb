```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1/(i) for i in range(1,7)])
plt.plot([2/(i+0.5) for i in range(1,7)])
plt.plot([3/(i+1) for i in range(1,7)])
plt.plot([4/(i+1.5) for i in range(1,7)])
```

```python
df = pd.DataFrame()
df["G50 - Humanoïde"] = [1/(i) for i in range(1,7)]
df["G20 - Humanoïde"] = [3/(i+1) for i in range(1,7)]
df["G50 - Non Humanoïde"] = [2/(i+0.5) for i in range(1,7)]
df["G20 - Non Humanoïde"] = [4/(i+1.5) for i in range(1,7)]
df.index = list(range(1,7))

fig, ax = plt.subplots()
ax.set_xlabel('Essai')  # Add an x-label to the axes.
ax.set_ylabel('Mesure du terrain commun')  # Add a y-label to the axes.
ax.set_yticks(ticks=np.linspace(0, 1.5, num=6), labels=[])
ax.set_xticks(ticks=range(1,7))
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
df.plot(ax=ax)
plt.savefig("res_attendu.png")
```
