```python
def graph(mu, sigma, nbtir):
    tirages = normal(mu, sigma, nbtir)

    statordre = np.sort(tirages)
    haut = np.arange(0,len(statordre)+1)/len(statordre)
    plt.hlines(haut, xmin=np.append(mu-4*sigma, statordre), xmax=np.append(statordre, mu+4*sigma), colors="red")
```

```python
graph(0,1,5)
```

```python
graph(0,1,20)
```

```python
def graph2(mu, sigma, dx):
    X = np.arange(np.floor(mu-4*sigma), np.ceil(mu+4*sigma), dx)
    Y = np.exp( -np.square(X-mu)/(2*sigma**2) ) /np.sqrt(2*np.pi) / sigma
    CY= np.cumsum(Y * dx)
    plt.plot(X, CY, color="blue")

graph2(mu, sigma, 1)
```

```python
graph2(1)
```

```python
def graph3(mu, sigma, nbtir, dx):
    tirages = normal(mu, sigma, nbtir)
    statordre = np.sort(tirages)
    haut = np.arange(0,len(statordre)+1)/len(statordre)
    plt.hlines(haut, xmin=np.append(mu-4*sigma, statordre), xmax=np.append(statordre, mu+4*sigma), colors="red")

    X = np.arange(np.floor(mu-4*sigma), np.ceil(mu+4*sigma), dx)
    Y = np.exp( -np.square(X-mu)/(2*sigma**2) ) /np.sqrt(2*np.pi) / sigma
    CY= np.cumsum(Y*dx)
    plt.plot(X, CY, color="blue")

graph3(0, 1, 40, 0.1)
```

```python
def tirageTemps(dx):
    X = np.arange(-20, 20, dx)
    Y = list()
    for t in X:
        if t<0:
            Y.append(0)
        elif t>=15:
            Y.append(1)
        else:
            Y.append((t+5)/20)
    plt.plot(X, Y, color="orange")

tirageTemps(0.01)
```
